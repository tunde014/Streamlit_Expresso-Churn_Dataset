```python
def drop(df):
    for i in df.columns:
        if ((df[i].isnull().sum()/len(df)) * 100) > 50:
            df.drop([i], axis = 1, inplace = True)
```

```python
def fill(df):
    for i in df.columns:
        if df[i].dtype in ['int64', 'float64']:
            df[i].fillna(df[i].median(), inplace = True)

        elif df[i].dtype == 'O':
            df[i].fillna(df[i].mode()[0], inplace = True)

    return df
```

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()

df1['REGION'] = lb.fit_transform(df1['REGION'])
df1['TENURE'] = lb.fit_transform(df1['TENURE'])
df1['MRG'] = lb.fit_transform(df1['MRG'])
df1['TOP_PACK'] = lb.fit_transform(df1['TOP_PACK'])
```

```python
df1.info()
```

```python
x = df1.drop('CHURN', axis = 1)
```

```python
y = df1['CHURN']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()
x_newscaled = mm.fit_transform(x)
x_newscaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train_scaled, x_test, y_train, y_test = train_test_split(x_newscaled,y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 19, max_features = 15, max_depth = None, min_samples_split = 10)
model.fit(x_train_scaled, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)

print(cm)
print(cr)
```

```python
from xgboost import XGBClassifier

model_xg = XGBClassifier( n_estimators =9, max_depth = 5, learning_rate = 1, objective = 'binary:logistic')

model_xg.fit(x_train_scaled, y_train)
```

```python
pred_xg = model_xg.predict(x_test)
pred_xg
```

```python
model_xg.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, pred_xg)
cr = classification_report(y_test, pred_xg)

print(cm)
print(cr)
```
